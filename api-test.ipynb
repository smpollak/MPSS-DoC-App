```python
import requests
import json
import pandas as pd
```

```python
def request_gini(code):
    url = f"https://api.worldbank.org/v2/country/{code}/indicator/SI.POV.GINI?format=json"
    url_text = requests.get(url).text
    gini_dict = json.loads(url_text)[1]
    rows = []
    for item in gini_dict:
        #print(item)
        if item['value'] != None:
            d = {}
            d['country'] = item['country']['value']
            d['year'] = item['date']
            d['GINI'] = item['value']
            rows.append(d)
    gini_df = pd.DataFrame(rows)

    return gini_df

request_gini('usa')
```
